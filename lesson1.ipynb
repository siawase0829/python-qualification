```python
import curses

def main(stdscr):
    # 端末の設定
    curses.curs_set(0)  # カーソルを非表示にする
    stdscr.nodelay(1)   # ノンブロッキングモードにする
    stdscr.timeout(100) # 100msごとにタイムアウト

    # 初期位置
    sh, sw = stdscr.getmaxyx()
    w = curses.newwin(sh, sw, 0, 0)
    turtle = '^'
    x, y = sh // 2, sw // 2

    # キー入力用の辞書
    key = curses.KEY_RIGHT

    while True:
        next_key = w.getch()
        key = key if next_key == -1 else next_key

        if key == ord('q'):
            break

        # 現在の亀の位置を消す
        w.clear()

        # 移動方向を決定
        if key == curses.KEY_UP:
            turtle = '^'
            x -= 1
        elif key == curses.KEY_DOWN:
            turtle = 'v'
            x += 1
        elif key == curses.KEY_LEFT:
            turtle = '<'
            y -= 1
        elif key == curses.KEY_RIGHT:
            turtle = '>'
            y += 1

        # 亀が端を越えないようにする
        x = max(0, min(x, sh - 1))
        y = max(0, min(y, sw - 1))

        # 新しい亀の位置を表示
        w.addch(x, y, turtle)
        w.refresh()

curses.wrapper(main)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# データ生成
x = np.linspace(-5, 5, 100)
y = np.sin(x)
z = np.cos(x)

# 3Dグラフ作成
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 3D線グラフをプロット
ax.plot(x, y, z)

# ラベル設定
ax.set_xlabel('X軸')
ax.set_ylabel('Y軸')
ax.set_zlabel('Z軸')

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# データ生成
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# 3Dグラフ作成
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 3Dサーフェスプロットを作成
ax.plot_surface(x, y, z, cmap='viridis')

# ラベル設定
ax.set_xlabel('X軸')
ax.set_ylabel('Y軸')
ax.set_zlabel('Z軸')

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import time

# データ生成
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# 3Dグラフ作成
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 3Dサーフェスプロットを作成
ax.plot_surface(x, y, z, cmap='viridis')

# 角度を変えて表示する
for angle in range(0, 360, 10):
    ax.view_init(elev=30, azim=angle)
    plt.draw()
    time.sleep(0.5)

```
