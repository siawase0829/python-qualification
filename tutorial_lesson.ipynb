```python
'''簡単な正弦波を作ってnp.fft.fft()を試す'''

'''必要なライブラリをインポート'''
import matplotlib.pyplot as plt
import numpy as np

'''超簡単な正弦波を作る'''
f = 10 # 周波数f(frequency),単位はHz(サイクル毎秒)
f_s = 100 # サンプリングレート。1秒あたりの測定数

t = np.linspace(0, 2, 2 * f_s, endpoint = False)# 時間を0から2秒に指定しサンプリング分割を行う(終点は要素に含まない)
x = np.sin(f * 2 * np.pi * t) # x=sin(2πf)の波形作成

'''グラフに表示'''
fig, ax= plt.subplots()
ax.plot(t, x) # 横軸にt, 縦軸にxの値をプロットする
ax.set_xlabel('Time[s]')
ax.set_ylabel('Signal amplitude')


'''フーリエ変換'''
X = np.fft.fft(x) # 波形のフーリエ変換
freqs = np.fft.fftfreq(len(x)) * f_s # サンプリング周波数を返す

'''グラフに表示'''
fig2, ax2= plt.subplots()
ax2.stem(freqs, np.abs(X), use_line_collection=True) # フーリエ変換の結果をステム(茎)プロットする
ax2.set_xlabel('Frequency in Hertz[Hz]')
ax2.set_ylabel('Frequency Domain (Spectrum) Magnitude')
ax2.set_xlim(- f_s / 2, f_s / 2)
ax2.set_ylim(-5, 110)

plt.show()
```

```python
'''fftを使って短時間フーリエ変換(stft)'''

'''必要なライブラリをインポート'''
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt
from skimage import util

'''waveファイルから必要な情報を読み込む'''
rate, audio = wavfile.read(r"音声データの置いてあるパス/音声ファイル名.wav", "r") # waveファイルを読み込み用ファイルとして開く

audio = np.mean(audio, axis=1) # ステレオをモノラルに変換
N = audio.shape[0] # 要素数を返す
L = N/rate # 長さの算出

'''一旦グラフにプロット'''
f, ax = plt.subplots()
ax.plot(np.arange(N)/rate, audio)
ax.set_xlabel('Time[s]')
ax.set_ylabel('Amplitude[unknown]')
plt.show()

'''
1024(N)個のサンプル値で構成され、1つ前のスライスとサンプル100(step)個分ずれるようなスライスに分割する
'''
M = 1024 # スライスのサンプル数

slices = util.view_as_windows(audio, window_shape=(M,), step=100) # スライス分割

window = np.hamming(M+1)[:-1] # 窓関数の生成
slices = slices * window # スライスに窓を掛け合わせる

'''列ごとに1つのスライスがあるほうが便利なのでデータを転置する'''
slices = slices.T

spectrum = np.fft.fft(slices, axis=0)[:N] # スライスをフーリエ変換
spectrum = np.abs(spectrum) # スペクトラムの絶対値を取っている

'''
スペクトルには、非常に大きい値と小さい値の両方が含まれるため、対数値を取って範囲を圧縮する
そしてグラフにプロットする
'''
f, ax = plt.subplots()

S = np.abs(spectrum) # スペクトラムの絶対値を取る
S = 20 * np.log10(S / np.max(S)) # スペクトラムの対数値を取る

ax.imshow(S, origin='lower', cmap='viridis', extent=(0, L, 0, rate/2/1000))
ax.axis('tight')
ax.set_ylabel('Frequency[kHz]')
ax.set_xlabel('Time[s]')

plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def inverse_kinematics(x, y, L1, L2):
    # Calculate the distance from the origin to the target point
    distance = math.sqrt(x**2 + y**2)

    # Check if the target is reachable
    if distance > (L1 + L2):
        raise ValueError("Target is out of reach")

    # Calculate the angle of the first joint (θ1)
    cos_theta2 = (x**2 + y**2 - L1**2 - L2**2) / (2 * L1 * L2)
    sin_theta2 = math.sqrt(1 - cos_theta2**2)  # Taking the positive root

    theta2 = math.atan2(sin_theta2, cos_theta2)

    k1 = L1 + L2 * cos_theta2
    k2 = L2 * sin_theta2

    theta1 = math.atan2(y, x) - math.atan2(k2, k1)

    return theta1, theta2

def forward_kinematics(theta1, theta2, L1, L2):
    # Calculate the position of the second joint
    x1 = L1 * math.cos(theta1)
    y1 = L1 * math.sin(theta1)

    # Calculate the position of the end effector
    x2 = x1 + L2 * math.cos(theta1 + theta2)
    y2 = y1 + L2 * math.sin(theta1 + theta2)

    return (0, 0), (x1, y1), (x2, y2)

# Link lengths
L1 = 1.0
L2 = 1.0

# Target position
target_x = 1.0
target_y = 1.0

# Solve inverse kinematics
try:
    theta1, theta2 = inverse_kinematics(target_x, target_y, L1, L2)
    print(f"Joint angles: θ1 = {math.degrees(theta1):.2f}°, θ2 = {math.degrees(theta2):.2f}°")

    # Solve forward kinematics to verify
    joint1, joint2, end_effector = forward_kinematics(theta1, theta2, L1, L2)
    print(f"End effector position: {end_effector}")

    # Plot the robot arm
    x_coords = [joint1[0], joint2[0], end_effector[0]]
    y_coords = [joint1[1], joint2[1], end_effector[1]]

    plt.figure()
    plt.plot(x_coords, y_coords, 'bo-', linewidth=2)
    plt.plot(target_x, target_y, 'rx')
    plt.xlim(-L1-L2-0.5, L1+L2+0.5)
    plt.ylim(-L1-L2-0.5, L1+L2+0.5)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('2-Link Robot Arm Inverse Kinematics')
    plt.grid()
    plt.show()
except ValueError as e:
    print(e)

```

```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# 定数
L = [1, 1, 1, 1, 1, 1]  # 各リンクの長さ

def forward_kinematics(thetas):
    """
    順運動学: 関節角度からエンドエフェクタの位置を計算
    """
    x = 0
    y = 0
    angle = 0
    
    for i in range(len(thetas)):
        angle += thetas[i]
        x += L[i] * np.cos(angle)
        y += L[i] * np.sin(angle)
    
    return np.array([x, y])

def inverse_kinematics(target, initial_guess):
    """
    逆運動学: ターゲット位置に対する関節角度を数値的に求める
    """
    def objective(thetas):
        effector_pos = forward_kinematics(thetas)
        return np.linalg.norm(effector_pos - target)
    
    result = minimize(objective, initial_guess, bounds=[(-np.pi, np.pi)]*6)
    return result.x if result.success else None

def plot_robot(thetas):
    """
    ロボットのプロット
    """
    x = 0
    y = 0
    angle = 0
    
    x_coords = [x]
    y_coords = [y]
    
    for i in range(len(thetas)):
        angle += thetas[i]
        x += L[i] * np.cos(angle)
        y += L[i] * np.sin(angle)
        x_coords.append(x)
        y_coords.append(y)
    
    plt.plot(x_coords, y_coords, 'bo-', linewidth=2)
    plt.xlim(-sum(L), sum(L))
    plt.ylim(-sum(L), sum(L))
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('6-Link Robot Arm')
    plt.grid()
    plt.show()

# ターゲット位置
target_position = np.array([3, 1])

# 初期推測
initial_guess = np.zeros(6)

# 逆運動学を解く
joint_angles = inverse_kinematics(target_position, initial_guess)

if joint_angles is not None:
    print("Joint angles:", joint_angles)
    plot_robot(joint_angles)
else:
    print("No solution found")
```

```python
class EachIterator:
    def __init__(self, iterable):
        self.iterable = iter(iterable)

    def __iter__(self):
        return self

    def __next__(self):
        return next(self.iterable)

def each(iterable):
    return EachIterator(iterable)

# 使用例
numbers = [1, 2, 3, 4, 5]

# Rubyのeachメソッドと同様の動作
for number in each(numbers):
    print(number)
```

```python
class EachIterator:
    def __init__(self, iterable):
        self.iterable = iter(iterable)

    def __iter__(self):
        return self

    def __next__(self):
        return next(self.iterable)
    
    @classmethod   
    def each(cls, func):
        for item in cls:
            func(item)



# 使用例
numbers = [1, 2, 3, 4, 5]
numbers.each = EachIterator.each

# list クラスに each メソッドを追加

# each メソッドを使ってイテレーション
numbers.each(print)
```

```python
def fibonacci(n):
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[i - 1] + fib[i - 2])
    return fib

n = 10  # 表示するフィボナッチ数列の長さ
fib_sequence = fibonacci(n)
print(f"フィボナッチ数列（{n}項目まで）: {', '.join(map(str, fib_sequence))}")
```

```python
import numpy as np

def inertia_matrix(q1, q2):
    # ロボットの質量行列
    m1, m2 = 1.0, 1.0
    l1, l2 = 1.0, 1.0
    I1, I2 = 1.0, 1.0
    
    M11 = m1 * l1**2 + m2 * (l1**2 + l2**2 + 2 * l1 * l2 * np.cos(q2)) + I1 + I2
    M12 = m2 * (l2**2 + l1 * l2 * np.cos(q2)) + I2
    M21 = M12
    M22 = m2 * l2**2 + I2
    
    M = np.array([[M11, M12],
                  [M21, M22]])
    return M

def coriolis_matrix(q1, q2, q1_dot, q2_dot):
    # コリオリ力と遠心力の項
    m2 = 1.0
    l1, l2 = 1.0, 1.0
    
    C11 = -m2 * l1 * l2 * np.sin(q2) * q2_dot
    C12 = -m2 * l1 * l2 * np.sin(q2) * (q1_dot + q2_dot)
    C21 = m2 * l1 * l2 * np.sin(q2) * q1_dot
    C22 = 0.0
    
    C = np.array([[C11, C12],
                  [C21, C22]])
    return C

def gravity_vector(q1, q2):
    # 重力項
    m1, m2 = 1.0, 1.0
    l1, l2 = 1.0, 1.0
    g = 9.81
    
    G1 = (m1 * l1 + m2 * l1) * g * np.cos(q1) + m2 * l2 * g * np.cos(q1 + q2)
    G2 = m2 * l2 * g * np.cos(q1 + q2)
    
    G = np.array([G1, G2])
    return G

def forward_dynamics(q, q_dot, tau):
    q1, q2 = q
    q1_dot, q2_dot = q_dot
    
    M = inertia_matrix(q1, q2)
    C = coriolis_matrix(q1, q2, q1_dot, q2_dot)
    G = gravity_vector(q1, q2)
    
    # 逆行列を使って加速度を計算
    q_ddot = np.linalg.inv(M).dot(tau - C.dot(q_dot) - G)
    return q_ddot

# 例として関節角度、角速度、トルクを与えて加速度を計算
q = [np.pi / 4, np.pi / 4]
q_dot = [0.1, 0.1]
tau = [1.0, 1.0]

q_ddot = forward_dynamics(q, q_dot, tau)
print("関節角加速度:", q_ddot)
```

```python
class MethodObject:
    def __init__(self, func):
        self.func = func

    def __get__(self, instance, owner):
        if instance is None:
            # クラスからのアクセス時にはそのままメソッドオブジェクトを返す
            return self
        else:
            # インスタンスからのアクセス時にはバインドメソッドを返す
            def bound_method(*args, **kwargs):
                # インスタンスメソッドとして呼び出すために、インスタンスを最初の引数として渡す
                return self.func(instance, *args, **kwargs)
            return bound_method

# メソッドオブジェクトの作成
def example_method(self):
    print(f"Method called on {self}")

method_obj = MethodObject(example_method)

# クラス定義
class MyClass:
    my_method = method_obj

# インスタンス化
obj = MyClass()

# メソッドオブジェクトの__get__を通じてインスタンスにバインド
bound_method = type(obj).__dict__["my_method"].__get__(obj, MyClass)

# バインドされたメソッドを呼び出す
bound_method()  # 出力:
obj.my_method()
```
