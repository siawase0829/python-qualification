```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# 伝達関数 G(s) = (s + 2) / (s^2 + 2s + 5)
numerator = [1, 2]  # 分子 (s + 2)
denominator = [1, 2, 5]  # 分母 (s^2 + 2s + 5)

# 伝達関数を定義
sys = signal.TransferFunction(numerator, denominator)

# 周波数応答を計算
w, mag, phase = signal.bode(sys)

# ボード線図 (周波数 vs 振幅)
plt.figure()
plt.subplot(2, 1, 1)
plt.semilogx(w, mag)  # 周波数軸を対数表示
plt.title('Bode plot')
plt.ylabel('Magnitude (dB)')
plt.grid(True)

plt.show()
# ボード線図 (周波数 vs 位相)
plt.subplot(2, 1, 2)
plt.semilogx(w, phase)
plt.ylabel('Phase (degrees)')
plt.xlabel('Frequency (rad/s)')
plt.grid(True)


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# 伝達関数 G(s) = (s + 2) / (s^2 + 2s + 5)
numerator = [1, 2]  # 分子 (s + 2)
denominator = [1, 2, 5]  # 分母 (s^2 + 2s + 5)

# 伝達関数を定義
sys = signal.TransferFunction(numerator, denominator)

# 周波数範囲を設定
w = np.logspace(-2, 2, 1000)  # 0.01から100までの対数刻み

# 周波数応答を取得
w, h = signal.freqresp(sys, w=w)

# ナイキスト線図を描画 (実部 vs 虚部)
plt.plot(h.real, h.imag, label='Nyquist plot')
plt.plot(h.real, -h.imag, label='Conjugate', linestyle='--')  # 対称部分も表示

# 原点 (0, 0) にクロスマークを追加
plt.plot(0, 0, 'ro')

# グラフの設定
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title('Nyquist plot')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.legend()

plt.grid()
plt.show()
```

```python
import numpy as np
from scipy import signal
import plotly.graph_objects as go

# サンプルデータの生成
t = np.linspace(0, 1, 500, endpoint=False)
x = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(500)

# バターワースフィルタを適用
b, a = signal.butter(3, 0.05)
y = signal.filtfilt(b, a, x)

# プロット
fig = go.Figure()

# 元のデータ
fig.add_trace(go.Scatter(x=t, y=x, mode='lines', name='Original Signal'))

# フィルタリング後のデータ
fig.add_trace(go.Scatter(x=t, y=y, mode='lines', name='Filtered Signal'))

# グラフのタイトルとラベルを設定
fig.update_layout(
    title='Signal Filtering with SciPy and Plotly',
    xaxis_title='Time',
    yaxis_title='Amplitude'
)

fig.show()
```

```python
from scipy import integrate

# 積分対象の関数 (例: f(x) = x^2)
def func(x):
    return x**2

# 積分範囲
x_vals = np.linspace(0, 10, 100)
y_vals = func(x_vals)

# SciPyを使って数値積分
result, error = integrate.quad(func, 0, 10)

# プロット
fig = go.Figure()

# 関数のプロット
fig.add_trace(go.Scatter(x=x_vals, y=y_vals, mode='lines', name='f(x) = x^2'))

# グラフのタイトルと積分結果を表示
fig.update_layout(
    title=f'Numerical Integration with SciPy (Result: {result:.2f})',
    xaxis_title='x',
    yaxis_title='f(x)'
)

fig.show()
```

```python
from scipy.stats import norm

# 正規分布のパラメータ
mu = 0  # 平均
sigma = 1  # 標準偏差

# データの生成
x = np.linspace(-5, 5, 500)
y = norm.pdf(x, mu, sigma)

# プロット
fig = go.Figure()

# 確率密度関数のプロット
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Normal Distribution PDF'))

# グラフのタイトルとラベルを設定
fig.update_layout(
    title='Normal Distribution PDF with SciPy and Plotly',
    xaxis_title='x',
    yaxis_title='Density'
)

fig.show()
```
